* Definir la función para extraer los textos

```python
import fitz

def get_text(url):
    # Get the text from a PDF file
    doc = fitz.open(url)

    # Extract text from each page and remove newline characters
    text = "".join([page.get_text().replace("\n", "") for page in doc])  
    
    return text


url_1 = "Data/Texto1.pdf"
url_2 = "Data/Texto2.pdf"
url_3 = "Data/Texto3.pdf"

text_1 = get_text(url_1)
text_2 = get_text(url_2)
text_3 = get_text(url_3)

print(text_1[:100])
```

```python
from NLP_Utilities.Tokenizer import Tokenizer

tokenizer = Tokenizer()

text_1_token = tokenizer.tokenize(text_1)

print(text_1_token)
```

```python
lemmas = "correr, corriendo, corrimos, corriamos, correriamos, niños, niño, niña, comidas, comer, comiendo, limpiando"

token = tokenizer.tokenize(lemmas)

print(token)
```

**For each text document we will obtain the co-ocurrence matriz so we will be able tu visualize the most defined concepts on the 3D Plot; in addition we will obtain the TF_IDF matrix to get the most relevant word. With this we can conclude which is the principal topic of the document with the help of the two metrics**

* For the **1°** document

```python
from NLP_Utilities.CoOccurrence_PCA import CoOccurrencePCA

co_ocurrence = CoOccurrencePCA()

# For the first text
co_ocurrence.fit_transform(docs=[text_1])

co_text_1 = co_ocurrence.compute_cooccurrence_matrix()

co_text_1
```

```python
co_ocurrence.apply_pca()

co_ocurrence.plot_3d_interactive()
```

```python
from NLP_Utilities.TF_IDF import TF_IDF

# For the first text
tf_idf = TF_IDF()

tf_idf.fit_transform(docs=[text_1])

tf_idf_1 = tf_idf.compute_tf_idf()

# Order the columns to get the most relevant words
tf_idf_1_numeric = tf_idf_1.select_dtypes(include='number')

tf_idf_1_sorted = tf_idf_1_numeric.sort_values(by=tf_idf_1_numeric.index[0], axis=1, ascending=True)

tf_idf_1_sorted
```

> We can assume that the principal topic for the first document is "navidad" and a story related to christmas
