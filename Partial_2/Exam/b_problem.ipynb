```python
from NLP_Utilities.CoOccurrence_PCA import CoOccurrencePCA
import fitz

def get_text(url):
    # Get the text from a PDF file
    doc = fitz.open(url)

    # Extract text from each page and remove newline characters
    text = "".join([page.get_text().replace("\n", "") for page in doc])

    return text

url_1 = "Data/Texto1.pdf"
url_2 = "Data/Texto2.pdf"
url_3 = "Data/Texto3.pdf"

text_1 = get_text(url_1)
text_2 = get_text(url_2)
text_3 = get_text(url_3)

print(text_1[:100])
```

```python
# Crear lista de documentos y nombres
documents = [text_1, text_2, text_3]
doc_names = ["Texto1", "Texto2", "Texto3"]

# Crear una matriz de co-ocurrencia para cada documento
cooccurrence_matrices = {}

for i, (doc, name) in enumerate(zip(documents, doc_names)):

    # Crear instancia para cada documento
    model = CoOccurrencePCA(window_size=2)

    # Procesar solo este documento
    model.fit_transform(docs=[doc], tittles=[name])

    # Calcular matriz de co-ocurrencia
    matrix = model.compute_cooccurrence_matrix()

    # Guardar la matriz
    cooccurrence_matrices[name] = matrix

    print(f"\n{name}:")
    print(f"  Vocabulario: {len(model.vocabulary)} palabras")
    print(f"  Matriz: {matrix.shape}")
    print(f"  Co-ocurrencias totales: {matrix.sum().sum()}")
```
